```python
# drawstyle
data = np.random.randn(30).cumsum()
data2 = np.random.randn(30).cumsum()
# figure(figsize=(12,6)) 인치단위의 크기 설정
fig = plt.figure()
fig = plt.figure(figsize=(12,6))
```

```python
# drawstyle
plt.plot(data2, 'k--', label='Default')
# default steps steps-pre steps-mid steps-post
plt.plot(data2, 'g-', drawstyle='steps-pre', label='Draw Style')
plt.plot(data2, 'r-', drawstyle='steps-mid', label='Draw Style')
plt.plot(data2, 'y-', drawstyle='steps-post', label='Draw Style')
# 다른 색상으로 pre, post
plt.plot(data2, 'b-', drawstyle='default', label='Draw Style')
plt.legend(loc='best')
```

```python
ratio = [30, 20, 10, 40]
explode = [0, 0, 0, 0.1]
labels = ['A', 'B', 'C', 'D']
colors = ['m', 'r', 'y', 'b']

graph = plt.figure(figsize=(5, 5))
sub1 = graph.add_subplot(111)

patches, texts, autotexts = sub1.pie(
    ratio,
    explode=explode,
    labels=labels,
    colors=colors,
    autopct="%1.1f%%",
    shadow=True,
    startangle=60)

for text in autotexts:
    text.set_fontsize(10)
    text.set_color('white')
```
